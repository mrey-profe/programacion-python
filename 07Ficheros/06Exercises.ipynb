# Exercises

## Exercise 1

Create a text file called `departments.txt` that contains a list of all the departments in a high school. Each department should be on a separate line.

Create a Python program that manages the teaching staff of a high school. The program should allow:

- Add a new teacher to a specific department. It should ask for their name, surname, and department (it should allow choosing from the available departments in the `departments.txt` file, listing them as a menu). Store them in the structure 
```json
{
    "English": [
        {
            "name": "María", 
            "surname": "López"
        },
        {
            "name": "Juan",
            "surname": "Pérez"
        } 
        ...
    ], 
    "Galician": [
        ...
    ], 
}
```
- Add a new department to the `departments.txt` file.

## Exercise 4

## Exercise 5

## Exercise 6

Create a Python program that checks if a file called `registrations.csv` exists with the following columns: `name`, `surnames`, `email` in the current directory. If it exists, it should read its content and generate the corresponding users in the system (you can simulate user creation by printing a message for each created user). The username should follow the format of the computer science domain users (a + year + name + surname initials, for example `a25mariarl` for María Rodríguez López registered in 2025). After processing the file, it should move it to a folder called `processed`, adding the current date and time to the file name (for example, `registrations_20240615_1030.csv`).
