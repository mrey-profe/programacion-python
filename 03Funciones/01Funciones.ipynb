# Funciones

## Para qu necesitamos funciones?

Hay 2 motivos principales por los que dividir el cdigo en funciones:

- Existe un determinado fragmento de cdigo que se repite muchas veces, tal cual o con pequeas variaciones.
- El cdigo se hace tan extenso que leerlo o comprenderlo se hace complicado (divide y vencers).

## Tipos de funciones (segn su origen)

- Funciones integradas en Python, por ejemplo, `print()`
- Funciones de los mdulos preinstalados, por ejemplo, `math`, `time`, etc.
- Funciones definidas por el programador.
- Mtodos de las clases (POO).

## Qu es una funcin

Una funcin es un bloque de cdigo que realiza alguna operacin. 

Puede recibir parmetros de entrada.

Puede devolver un valor como salida.



## Estructura de una funcin

```python
def nombre_de_la_funcion():
    cuerpo_de_la_funcion
```

- Siempre comienza con la palabra reservada `def` (que significa definir).
- Despus el nombre de la funcin (mismas reglas que para las variables).
- Despus del nombre de la funcin, hay un espacio para un par de parntesis (donde se definirn los argumentos a recibir por la funcin).
- La lnea debe de terminar con dos puntos.
- La lnea inmediatamente despus de def marca el comienzo del cuerpo de la funcin - donde varias o (al menos una) instruccin anidada ser - ejecutada cada vez que la funcin sea invocada; nota: la funcin termina donde el anidamiento termina.




````python
def saludar():
    print("Hola")

print("Antes de llamar a saludar")
saludar()
print("Despus de llamar a saludar una vez")
saludar()
print("Despus de llamar a saludar una segunda vez")
````

## Funciones parametrizadas

Las funciones pueden recibir parmetros

### Parmetros

Un parmetro es una variable, pero existen dos factores que hacen a un parmetro diferente:

- Los parmetros solo existen dentro de las funciones en donde han sido definidos, y el nico lugar donde un parmetro puede ser definido es entre los parntesis despus del nombre de la funcin, donde se encuentra la palabra reservada def.
- La asignacin de un valor a un parmetro de una funcin se hace en el momento en que la funcin se invoca, especificando el argumento correspondiente.

Si se define un parmetro en la definicin de la funcin, debe asignrsele un valor a ese parmetro cuando se invoque.

### Parmetros vs. argumentos

- Los parmetros solo existen dentro de las funciones (este es su entorno natural).
- Los argumentos existen fuera de las funciones, y son los que pasan los valores a los parmetros correspondientes.

````python
def saludar(nombre):
    print("Hola,", nombre)

print("Antes de llamar a saludar")
saludar("Ana")
print("Despus de llamar a saludar una vez")
saludar("Xurxo")
print("Despus de llamar a saludar una segunda vez")
````

Los argumentos no tienen necesariamente que ser literales, pueden ser variables:

## Argumentos posicionales

Los argumentos se asignan a los parmetros en el orden en el que se indican.

````python
def presentacion(nombre, apellido):
    print("Hola, me llamo", nombre, apellido)

presentacion("Ana", "Garca")
````

## Argumentos de palabra clave

El significado del argumento est definido por su nombre, no su posicin.

````python
def presentacion(nombre, apellido):
    print("Hola, me llamo", nombre, apellido)

presentacion(nombre="Ana", apellido="Garca")
presentacion(apellido="Rodrguez", nombre="Xin")
````

## Combinar argumentos posicionales y de palabra clave

Es posible combinar ambos tipos si se desea, solo hay una regla inquebrantable: **se deben colocar primero los argumentos posicionales y despus los de palabra clave**.

````python
def imprimir_volumen_prisma(ancho, largo, altura):
    volumen = ancho * largo * altura
    print(volumen)

imprimir_volumen_prisma(2, altura = 3, largo = 4)
# Escribe 2 llamadas ms con los mismos valores

imprimir_volumen_prisma(altura = 3, largo = 4, 2)
````

## Devolucin de resultado con `return`

La instruccin `return` (palabra reservada) nos sirve para salir de una funcin, devolviendo o no un resultado.

### `return` sin una expresin

La palabra reservada en s, sin nada que la siga.

Cuando se emplea dentro de una funcin, provoca la terminacin inmediata de la ejecucin de la funcin, y un retorno instantneo (de ah el nombre) al punto de invocacin.

````python
def felicitar_ao(deseos = True):
    print("Tres...")
    print("Dos...")
    print("Uno...")
    if not deseos:
        return
    
    print("Feliz ao nuevo!")

felicitar_ao()
print()
felicitar_ao(False)
````


### `return` con una expresin

Va seguida de una expresin, que contiene el resultado de la funcin. Esta expresin puede ser tan sencilla como un literal o una variable o cualquier expresin ms compleja.

Tambin provoca la terminacin inmediata de la ejecucin de la funcin, y un retorno instantneo (de ah el nombre) al punto de invocacin.

````python
def area_triangulo(base, altura):
    return base * altura / 2

nombre = input("Escribe tu nombre: ")
base = int(input("Introduce la base del tringulo: "))
altura = int(input("Introduce la altura del tringulo: "))
#print("El rea del tringulo es", area_triangulo(base, altura))
print("El rea del tringulo es", area_triangulo(base=base, altura=altura))
````

## Devolucin de varios valores

En Python (lo que no es habitual en otros lenguajes) es posible que las funciones devuelvan varios valores. Para ello, solo tenemos que separarlos con una coma en el `return` y almacenarlos en dos variables separadas por una coma al llamar a la funcin (en este caso se asignan por orden).

````python
def area_perimetro_rectangulo(base, altura):
    area = base * altura
    perimetro = 2 * (base + altura)
    return area, perimetro

a, p = area_perimetro_rectangulo(3, 5)
print("El rea es", a, "y el permetro es", p)
````

Tambin se pueden asignar las variables de dos en dos:

````python
x, y = 3, 4

print("x = ", x, "y = ", y)
````

E incluso intercambiarlas:

````python
x = 3
y = 5

x, y = y, x

print("x = ", x, "y = ", y)
````

Y asignacin mltiple:

````python
a = 1
b = 2
c = 3

a, b, c = c, a, b # OJO! Esto no es un intercambio de valores, es una asignacin mltiple
# Equivale a:
# a = c
# b = a
# c = b

print("a = ", a, "b = ", b, "c = ", c)
````

- Cuando se le asigna a una variable (o se devuelve como el resultado de una funcin).
- Cuando se compara con una variable para diagnosticar su estado interno.



````python
value = None
if value is None: # if value == None:
    print("Lo siento, no contienes ningn valor")
````

 Si una funcin no devuelve un cierto valor utilizando una clusula de expresin return, se asume que devuelve implcitamente None.

````python
def strange_function(n):
    if(n % 2 == 0):
        return True
    
print(strange_function(2))
print(strange_function(1))
````
