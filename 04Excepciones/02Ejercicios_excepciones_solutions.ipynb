## Exercise 1

Modify Exercise 4 (show menu) from Functions to check if the user introduces an integer. If not, it should display an error.

```python
def menu():
    print('''1. Greet
    2. Say goodbye
    3. Exit
    Choose an option:''')
    try:
        resultado = int(input())
        return resultado
    except ValueError:
        print("Iso non é un enteiro")

def operation(n):
    match n:
        case 1:
            print("Hello")
        case 2:
            print("Goodbye")
        case 3:
            print("Exit")


option = menu()
if option is not None:
    operation(option)
```

## Exercise 2
Modify Exercise 1 to keep asking the user until they write an integer.

```python
def menu():
    print('''1. Greet
    2. Say goodbye
    3. Exit
    Choose an option:''')
    valor_correcto = False
    while not valor_correcto:
        try:
            resultado = int(input())
            valor_correcto =  True
        except ValueError:
            print("Iso non é un enteiro")
    return resultado

def operation(n):
    match n:
        case 1:
            print("Hello")
        case 2:
            print("Goodbye")
        case 3:
            print("Exit")


option = menu()
if option is not None:
    operation(option)
```

## Exercise 3
Using exception handling, create a program that asks the user for a number and displays the square root of that number on the screen. If the number is negative, a message should be shown indicating that it is not possible to calculate the square root of a negative number.

```python
import math 

try:
    number = float(input("Write a number:"))
    try:
        print(math.sqrt(number))
    except ValueError:
        print("This operation is not allowed for real numbers")
except ValueError:
    print("This is not a real number")
```

## Exercise 4
Modify the previous program so that it keeps asking the user for a value until they enter a positive number.

```python
import math 

correct_number = False
while not correct_number:
    try:
        number = float(input("Write a number:"))
        try:
            print(math.sqrt(number))
            correct_number = True
        except ValueError:
            print("This operation is not allowed for real numbers")
    except ValueError:
        print("This is not a real number")
```

## Exercise 5
Modify Exercise 7 ADVANCED from Functions (leap years between max and min) to use the default max year if the user types RETURN instead of an integer.

```python
# ADVANCED version 1
def is_leap_year(year: int) -> bool:
    '''
    Recibe un año y determina si es bisiesto o no.
    
    Parámetros
    ----------
    year: int
        Año a evaluar.
        
    Devuelve
    -------
    bool
        True si el año es bisiesto, False en caso contrario.
        
    >>> is_leap_year(2000)
    True
    >>> is_leap_year(1900)
    False
    >>> is_leap_year(1996)
    True
    >>> is_leap_year(1997)
    False
    '''
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        return True
    return False

def leap_years(start: int, end: int = 2025) -> None:
    for year in range(start, end + 1):
        if is_leap_year(year):
            print(year)

start = int(input("Write the start year:"))
try:
    end = int(input("Write the end year: (ENTER for 2025)"))
    leap_years(start, end)
except ValueError:
    leap_years(start)
```
