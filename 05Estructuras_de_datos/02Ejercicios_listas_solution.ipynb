## Exercise 11

Look up the algorithm to calculate the Spanish ID card letter (DNI) and create a function that uses lists to return the letter.

````python
def calculate_dni_letter(dni_number: int) -> str:
    """Calculate the letter of a Spanish DNI given its number.
    
    Args:
        dni_number (int): The number of the DNI.
    Returns:
        str: The corresponding letter.
    >>> calculate_dni_letter(12345678)
    'Z'
    >>> calculate_dni_letter(87654321)
    'X'
    >>> calculate_dni_letter(0)
    'T'
    """
    letters = ["T", "R", "W", "A", "G", "M", "Y", "F", "P", "D", "X", "B", "N", "J", "Z", "S", "Q", "V", "H", "L", "C", "K", "E"]
    index = dni_number % 23
    return letters[index]
````

## Exercise 16

Create a 3x3 two-dimensional list to play tic-tac-toe. It will store 3 possible strings: an empty string to indicate that no piece has been placed in that position, an 'o' to indicate that player 1 has played, and an 'x' for player 2.

Create a function that traverses it and checks if player 1 has achieved tic-tac-toe in any of the rows.

Create a function that traverses it and checks if player 1 has achieved tic-tac-toe in any of the columns.

**ADVANCED**: Create a function that traverses it and checks if player 1 has achieved tic-tac-toe in any of the diagonals.

**ADVANCED**: Create a function that traverses it and checks if any player has achieved tic-tac-toe.

**ADVANCED**: Complete the exercise so that two players can play tic-tac-toe.

````python
# Tic-tac-toe helpers
from typing import List


def has_row_win(board: List[List[str]], player: str) -> bool:
    for row in board:
        win = True
        for cell in row:
            if cell != player:
                win = False
                break
        if win:
            return True
    return False

def has_col_win(board: List[List[str]], player: str) -> bool:
    size = len(board)
    for col in range(size):
        win = True
        for row in range(size):
            if board[row][col] != player:
                win = False
                break
        if win:
            return True
    return False

def has_diag_win(board: List[List[str]], player: str) -> bool:
    size = len(board)
    
    # Check main diagonal
    win = True
    for i in range(size):
        if board[i][i] != player:
            win = False
            break
    if win:
        return True
    
    # Check anti-diagonal
    win = True
    for i in range(size):
        if board[i][size - 1 - i] != player:
            win = False
            break
    if win:
        return True
    
    return False

def winner(board: List[List[str]]) -> str | None:
    for player in ("o", "x"):
        if has_row_win(board, player) or has_col_win(board, player) or has_diag_win(board, player):
            return player
    return None

# Example usage
board_example: List[List[str]] = [
    ["o", "x", "o"],
    ["x", "o", ""],
    ["o", "", "x"],
]

print("Row win o:", has_row_win(board_example, "o"))
print("Col win o:", has_col_win(board_example, "o"))
print("Diag win o:", has_diag_win(board_example, "o"))
print("Winner:", winner(board_example))
````

## Exercise 24

Create a list with the years that are leap years from a list of years.

For example, given [1995, 2004, 2003, 1900, 2008] it should return [2004, 2008].

You must use the `is_leap_year(year)` function created in the previous topic and list comprehension.

````python
def is_leap_year(year: int) -> bool:
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        return True
    return False

years = [1995, 2004, 2003, 1900, 2008]

leap_years = [year for year in years if is_leap_year(year)]

print(leap_years)
````
