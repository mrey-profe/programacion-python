## Exercise 1

Write a program that asks the user for their name, age, address, and phone number and stores it in a dictionary. Then it should display the message

`<name> is <age> years old, lives at <address> and their phone number is <phone>.`

## Exercise 2

Write a program that reads a string and returns a dictionary with the number of occurrences of each vowel in the string.

````python
text = input("Enter a string: ").lower()

vowels = "aeiou"
vowel_count = {}

for char in text:
    if char in vowels:
        if char in vowel_count:
            vowel_count[char] += 1 # If the vowel is already in the dictionary, increment its count
        else: # The vowel is not in the dictionary yet
            vowel_count[char] = 1

print("Vowel occurrences:", vowel_count)
````

## Exercise 3

Write a program that counts the frequency of words in a given text string. The program should return a dictionary where the keys are the words and the values are the frequency of each word.

You can complete this very similar Codewars exercise: https://www.codewars.com/kata/578de3801499359921000130


````python
text = input("Enter a text string: ")

# Split the text into words and convert to lowercase
words = text.lower().split()

# Count word frequency
word_frequency = {}
for word in words:
    # Remove punctuation from the word, punctuation is placed at the start or end of words
    clean_word = word.strip(".,!?;:")
    if clean_word:
        if clean_word in word_frequency:
            word_frequency[clean_word] += 1
        else:
            word_frequency[clean_word] = 1

print("Word frequency:", word_frequency)
````

## Exercise 4

Write a program that takes two tuples, one with names and another with ages, and converts them into a dictionary where the keys are the names and the values are the ages.

For example, if the input tuples are `("Alice", "Bob", "Charlie")` and `(25, 30, 35)`, the output should be `{"Alice": 25, "Bob": 30, "Charlie": 35}`.

````python
names = ("Alice", "Bob", "Charlie")
ages = (25, 30, 35)

dict_ages = {}

if len(names) != len(ages):
    print("Error: The number of names and ages must be the same.")
else:
    for i in range(len(names)):
        dict_ages[names[i]] = ages[i]
    
print("Dictionary:", dict_ages)
````

## Exercise 5

Write a program that uses a dictionary as a shopping list. The key will be the product name and the value will be the quantity. The program will ask for products and the quantity to be purchased, and will add them to the dictionary.

The program will display the shopping list.

```python
Product     Quantity
Apples      2
Pears       3
Oranges     1
```

Add a new dictionary that relates the product to its unit price.

Your program should use both dictionaries to display the shopping receipt with the total price for each product and the total price of the purchase.

```python
Product     Quantity    Unit price         Total price
Apples      2           0.35 €             0.7 €
Pears       3           0.40 €             1.2 €
Oranges     1           0.50 €             0.5 €

TOTAL                                      2.4 € 
```

````python
shopping_list = {}
prices = {
    "Apples": 0.35,
    "Pears": 0.40,
    "Oranges": 0.50,
    "Bananas": 0.30,
    "Milk": 1.20,
    "Bread": 0.90,
    "Eggs": 2.50,
    "Cheese": 3.80,
    "Chicken": 5.50,
    "Rice": 1.10
}

# Ask user for products
print("Enter products for your shopping list (enter 'done' when finished):")
product = input("Product name: ")

while product.lower() != "done":
    if product in prices:
        quantity = int(input(f"Quantity of {product}: "))
        shopping_list[product] = quantity
    else:
        print(f"Sorry, {product} is not available in our price list.")
    
    product = input("Product name (or 'done' to finish): ")

# Display shopping list
print(f"{'Product':<15} {'Quantity':<10}")
for product, quantity in shopping_list.items():
    print(f"{product:<15} {quantity:<10}")

# Display shopping receipt with prices
print(f"\n{'Product':<15} {'Quantity':<10} {'Unit price':<15} {'Total price':<15}")

total = 0
for product, quantity in shopping_list.items():
    unit_price = prices[product]
    total_price = quantity * unit_price
    total += total_price
    print(f"{product:<15} {quantity:<10} {unit_price:.2f} €{'':<12} {total_price:.2f} €")

print("\n" + f"{'TOTAL':<55} {total:.2f} €")
````

## Exercise 6 (exercise from previous year exam)

In preparation for the New Year's Eve dinner, a family stores the different menu options in a tuple and asks the dinner attendees which one they prefer. Create a Python program that:

- Declares that tuple and displays a menu asking attendees what their preferred option is (this should be repeated until it is indicated that there are no more people to decide).
- Stores the attendees' choices in a one-dimensional list (they can be repeated).
- Creates a function that receives the previously created list and returns a dictionary with the number of times each option was chosen.

````python
def count_choices(choices_list):
    """
    Receives a list of choices and returns a dictionary with the count of each option.
    
    Args:
        choices_list: List containing the selected options
        
    Returns:
        Dictionary with option counts
    """
    choice_count = {}
    for choice in choices_list:
        if choice in choice_count:
            choice_count[choice] += 1
        else:
            choice_count[choice] = 1
    return choice_count


# Menu options
menu_options = ("Turkey with chestnuts", "Sea bass", "Vegetarian lasagna", "Roast lamb")

# Display menu
print("New Year's Eve Dinner Menu:")
for i in range(len(menu_options)):
    print(f"{i + 1}. {menu_options[i]}")

# Store attendees' choices
choices = []

choice = 0
while choice != -1:
    print("\nMenu options:")
    for i in range(len(menu_options)):
        print(f"{i + 1}. {menu_options[i]}")
    print("-1. Finish")
    
    choice = int(input("Enter the option number (1-4) or -1 to finish: "))
    
    if choice != -1:
        if 1 <= choice <= len(menu_options):
            choices.append(menu_options[choice - 1])
            print(f"Added: {menu_options[choice - 1]}")
        else:
            print("Invalid option!")

# Count choices
choice_summary = count_choices(choices)

# Display results
print("\n" + "=" * 50)
print("SUMMARY OF CHOICES:")
for option, count in choice_summary.items():
    print(f"{option}: {count} person(s)")
print(f"Total attendees: {len(choices)}")
````
