## Exercise 1

Using a tuple, write a function that receives a month number and returns its name.

Create a main program to test it.

```python
def get_month_name(month_number):
    """Return the name of the month given its number (1-12)."""
    
    months = ("January", "February", "March", "April", "May", "June",
            "July", "August", "September", "October", "November", "December")
    if 1 <= month_number <= 12:
        return months[month_number - 1]
    else:
        return "Incorrect month number"
    
number = int(input("Enter month number (1-12): "))
print(get_month_name(number))
```

## Exercise 2

Look up the algorithm to calculate the Spanish ID (DNI) letter and create a function that uses tuples to return the corresponding letter.

```python
def calculate_dni_letter(dni_number: int) -> str:
    """Calculate the letter of a Spanish DNI given its number.
    
    Args:
        dni_number (int): The number of the DNI.
    Returns:
        str: The corresponding letter.
    >>> calculate_dni_letter(12345678)
    'Z'
    >>> calculate_dni_letter(87654321)
    'X'
    >>> calculate_dni_letter(0)
    'T'
    """
    letters = ("T", "R", "W", "A", "G", "M", "Y", "F", "P", "D", "X", "B", "N", "J", "Z", "S", "Q", "V", "H", "L", "C", "K", "E")
    index = dni_number % 23
    return letters[index]
```

## Exercise 3

Create a function that receives a day of the week and returns the next one, for example, if we pass "Monday" it will return "Tuesday".

```python
def next_day(current_day: str) -> str:
    """Return the next day of the week given the current day.
    
    Args:
        current_day (str): The current day of the week.
    Returns:
        str: The next day of the week.
    >>> next_day("Monday")
    'Tuesday'
    >>> next_day("Sunday")
    'Monday'
    """
    days_of_the_week = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")
    if current_day in days_of_the_week:
        current_index = days_of_the_week.index(current_day)
        next_index = (current_index + 1) % len(days_of_the_week)
        return days_of_the_week[next_index]
    else:
        return "Invalid day"
```

## Exercise 4: ADVANCED

In mathematics, the imaginary number $i$ is $\sqrt{-1}$. Thus its powers will be:

$i^0 = 1$

$i^1 = i$

$i^2 = -1$

$i^3 = -i$

$i^4 = 1 $

$i^5 = i $

$...$

Write a function that receives the exponent for a power of $i$ and returns the result as a string.

[Exercise on codewars](https://www.codewars.com/kata/5a97387e5ee396e70a00016d/train/python)

```python
def pow_complex_i(n: int) -> str:
    """Return the value of i raised to the power of n.
    
    Args:
        n (int): The exponent.
    Returns:
        complex: The result of i^n.
    >>> pow_complex_i(0)
    1
    >>> pow_complex_i(1)
    1j
    >>> pow_complex_i(2)
    (-1+0j)
    >>> pow_complex_i(3)
    -1j
    >>> pow_complex_i(4)
    1
    """
    powers_of_i = ("1", "i", "-1", "-i")
    index = n % 4
    return powers_of_i[index]
```
