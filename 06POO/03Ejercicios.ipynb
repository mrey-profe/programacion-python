# Ejercicio 6

Créela dos clases, la clase Punto y la clase Línea que se describen a continuación: 

Crea una clase `Punto` que modele un punto en un espacio bidimensional con las siguientes características: 

Tendrá dos atributos, x e y, que guardan las coordenadas. Los valores de estos atributos serán enteros entre -100 y 100. Solamente pueden ser accesibles desde la propia clase, ya que habrá métodos para obtener sus valores. 

Habrá otro constructor a lo que se le pasan las dos coordenadas del punto. Si los datos pasados son incorrectos (o no se le indican) creará un punto en (0,0).

Tendrá los siguientes métodos: 

- moverHorizontal(boolean positivo, int valor): modificará la coordenada x de la siguiente manera: 

    - Si positivo es true, incrementará su posición a lo que se indique en valor (siendo 100 como máximo) 

    - Si positivo es false, decrementará su posición a lo que se indique en valor(siendo -100 como mínimo) 

- moverVertical(boolean positivo, int valor): modificará la coordenada y de la siguiente manera: 

    - Si positivo es true, incrementará su posición a lo que se indique en valor (siendo 100 como máximo) 

    - Si positivo es false, decrementará su posición a lo que se indique en valor (siendo -100 como mínimo) 

- byte obtenerCuadrante(): devolverá un valor entre 1 y 4 (cada número representa a un cuadrante) en base a las siguientes condiciones: 

    - Si x>0 y y>0 el punto se encuentra en el cuadrante 1. Se devuelve un 1. 

    - Si x<0 y y>0 el punto se encuentra en el cuadrante 2. Se devuelve un 2. 

    - Si x<0 y y<0 el punto se encuentra en el cuadrante 3. Se devuelve un 3.  

    - Si x>0 y y<0 el punto se encuentra en el cuadrante 4. Se devuelve un 4. 

Crear el método __str__ para que devuelva el punto de la siguiente manera: “[valorX, valorY]” 


Crea la clase `Línea` que tendrá un array llamado coordenadas que almacenará dos Puntos. Estos dos puntos son por los que pasa la línea en un espacio de dos dimensiones. 

La clase dispondrá de los siguientes métodos: 

- Constructor que recibe dos puntos. Si los puntos no son válidos, se creará una línea con dos puntos en (0,0).

- boolean moverHorizontal (boolean positivo, int valor): mueve los dos puntos a la derecha (se positivo es true) o a la izquierda(se positivo es false) en la distancia que se indique en valor. Para moverlos, ningún punto puede salirse del plano (valor máximo 100 y mínimo -100). Se devolverá un booleano indicando se se hizo o no el movimiento. El movimiento pudo no hacerse porque valor no es un número positivo > 0 o porque la línea se sale del espacio del plano. 

- `boolean moverVertical (boolean positivo, int valor)`: mueve los dos puntos hacia arriba (se positivo es true) o hacia abajo (se positivo es false) en la distancia que se indique en valor. Para moverlos, ningún punto puede salirse del plano (valor máximo 100 y mínimo -100). Se devolverá un booleano indicando se se hizo o no el movimiento. El movimiento pudo no hacerse porque valor no es un número positivo > 0 o porque la línea se sale del espacio del plano. 

Sobreescribir el método __str__ para que devuelva línea de la siguiente manera: `“Punto La [valorX, valorY], Punto B [valorX, valorY]”`
